```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Mobile-Price-Prediction-cleaned_data.csv")
```

```python
df.head(10)
```

```python
df.drop(["Ratings"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
X = df.drop(["Price"], axis = 1)
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 0)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score 
```

```python
models ={
    "lr":LinearRegression(),
    "lasso": Lasso(),
    "rdge": Ridge(),
    "dtr": DecisionTreeRegressor(),
    "knn":KNeighborsRegressor(),
    "rfr": RandomForestRegressor()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    rmse = mse ** 0.5
    r2 = r2_score(y_test, y_pred)

    print(f"Name:{name}")
    print(f"R-squared:{rmse:.2f}")
    print(f"R2:{r2:.2f}")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

for name, model in models.items():
    model.fit(X_train_sc, y_train )
    y_pred_sc = model.predict(X_test_sc)
    
    mse_sc = mean_squared_error(y_test, y_pred_sc)
    rmse_sc = mse_sc ** 0.5
    r2_score_sc = r2_score(y_test, y_pred_sc)
    
    print(f"Name:{name}")
    print(f"R-squared:{rmse_sc:.2f}")
    print(f"R2:{r2_score_sc:.2f}")
```

```python
Data = {
    "id": np.arange(1, 243),
    "Actual Prediction": y_test,
    "Machine Prediction": y_pred_sc,
}
Result = pd.DataFrame(Data)
```
